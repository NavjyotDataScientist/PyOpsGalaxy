````python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Step 1: Email account details
sender_email = "your_email@gmail.com"
receiver_emails = ["team_member1@company.com", "team_member2@company.com"]  # Add all team members
password = "your_app_password"  # Use App Passwords for Gmail/Outlook

# Step 2: Create the email
message = MIMEMultipart()
message["From"] = sender_email
message["To"] = ", ".join(receiver_emails)
message["Subject"] = "Request for Daily Reports"

# Draft email body with signature
body = """
Dear Team,

I hope this message finds you well.
As part of improving our workflow, I kindly request each of you to send me your **daily report** by the end of the working day.
This will help us track progress more effectively and avoid last-minute rushes.

Thank you for your cooperation and commitment.

Best regards,
Your Name
Software Engineer | Company Name
üìû +91-XXXXXXXXXX
‚úâÔ∏è your_email@gmail.com
"""

message.attach(MIMEText(body, "plain"))

# Step 3: Connect to Gmail SMTP server and send
try:
    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_emails, message.as_string())
    print("‚úÖ Email sent successfully!")
except Exception as e:
    print("‚ùå Error:", e)
finally:
    server.quit()
````

### Overview of the script

This Python script is designed to send a standardized email requesting daily reports to a list of team members using an SMTP server (in this case, Gmail's).

### Core Libraries Used

-   `smtplib`: This module defines an SMTP client session object that can be used to send mail to any Internet machine with an SMTP or ESMTP listener daemon.
-   `email.mime.text.MIMEText`: This class is used to create email messages that are plain text.
-   `email.mime.multipart.MIMEMultipart`: This class is used to create email messages that combine different parts (like text and attachments, although here it's just text) into a single message.

### Step 1: Email Account Details

```python
# Step 1: Email account details
sender_email = "your_email@gmail.com"
receiver_emails = ["team_member1@company.com", "team_member2@company.com"]  # Add all team members
password = "your_app_password"  # Use App Passwords for Gmail/Outlook
```

-   `sender_email`: This is your email address from which the email will be sent.
-   `receiver_emails`: This is a list of email addresses of the recipients. You can add as many team members as needed.
-   `password`: **Crucially, for services like Gmail, you should use an "App Password" instead of your regular account password, especially if you have 2-Factor Authentication enabled.** An App Password is a 16-digit passcode that gives an app or device permission to access your Google Account. This is a more secure way to grant access to third-party applications. To generate one for Gmail, you typically go to your Google Account Security settings -> App Passwords.

### Step 2: Create the Email

```python
# Step 2: Create the email
message = MIMEMultipart()
message["From"] = sender_email
message["To"] = ", ".join(receiver_emails)
message["Subject"] = "Request for Daily Reports"

# Draft email body with signature
body = """
Dear Team,

I hope this message finds you well.  
As part of improving our workflow, I kindly request each of you to send me your **daily report** by the end of the working day.  
This will help us track progress more effectively and avoid last-minute rushes.

Thank you for your cooperation and commitment.

Best regards,  
Your Name  
Software Engineer | Company Name  
üìû +91-XXXXXXXXXX  
‚úâÔ∏è your_email@gmail.com
"""

message.attach(MIMEText(body, "plain"))
```

-   `message = MIMEMultipart()`: Initializes a multipart email message object. This allows for more complex email structures (e.g., combining plain text with HTML, or adding attachments).
-   `message["From"]`, `message["To"]`, `message["Subject"]`: These lines set the sender, recipients, and subject of the email, respectively. Note that `message["To"]` joins the list of receiver emails into a single comma-separated string.
-   `body`: This multiline string contains the actual content of your email. It includes placeholders like `Your Name` and contact details that you would replace with your own information.
-   `message.attach(MIMEText(body, "plain"))`: This attaches the plain text `body` to the `MIMEMultipart` message object. The second argument, `"plain"`, specifies the content type.

### Step 3: Connect to SMTP Server and Send

```python
# Step 3: Connect to Gmail SMTP server and send
try:
    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_emails, message.as_string())
    print("‚úÖ Email sent successfully!")
except Exception as e:
    print("‚ùå Error:", e)
finally:
    server.quit()
```

-   `try...except...finally`: This block is used for error handling:
    -   `server = smtplib.SMTP("smtp.gmail.com", 587)`: Establishes a connection to Gmail's SMTP server. `"smtp.gmail.com"` is the server address, and `587` is the standard port for secure SMTP.
    -   `server.starttls()`: This command upgrades the connection to TLS (Transport Layer Security) encryption. It's crucial for secure communication over the internet.
    -   `server.login(sender_email, password)`: Authenticates your `sender_email` with the `password` (your App Password).
    -   `server.sendmail(sender_email, receiver_emails, message.as_string())`: Sends the email. It takes the sender's email, a list of receiver emails, and the email message itself (converted to a string using `message.as_string()`).
    -   `print("‚úÖ Email sent successfully!")`: A success message if the email is sent without issues.
    -   `except Exception as e: print("‚ùå Error:", e)`: Catches any exceptions (errors) that occur during the process and prints an error message. A common error here is incorrect `sender_email` or `password` (e.g., if you don't use an App Password).
    -   `finally: server.quit()`: Ensures that the connection to the SMTP server is properly closed, regardless of whether the email was sent successfully or an error occurred. This is good practice to release resources.
